### 📊 Nigeria Financial Market Dataset Analysis

#### Overview
The Nigeria Financial Exchange Market dataset downloaded from CBN site contains financial market data capturing exchange rates and turnover activity across different trading segments. It provides insight into market liquidity, pricing trends, and trading volume dynamics.

#### Key Variables
- **ratedate**: The date of the recorded market data
- **closingrate**: The closing exchange rate for the day
- **nfeM_Total_Turnover**: Total turnover in the NFEM market segment
- **interBank_Total_Turnover**: Total turnover in the interbank market

#### Objective
The goal of this analysis is to:
- Examine trends in exchange rates over time  
- Analyze liquidity patterns using turnover data  
- Identify anomalies or significant movements  
- Generate actionable financial insights  

#### Scope
This analysis focuses on:
- Time-based trends  
- Market activity comparison  
- Statistical summaries  
- Insight generation using AI (Qwen model)

```python
import os
from reportlab.platypus import SimpleDocTemplate, Paragraph, Spacer
from reportlab.lib.styles import getSampleStyleSheet
from reportlab.lib.pagesizes import letter

def create_pdf(summary_stats, ai_text):
    # 1. Define the exact folder path
    folder = r"C:\Users\ACER\Desktop\Portfolios\NFEM"
    
    # Create folder if it doesn't exist
    if not os.path.exists(folder):
        os.makedirs(folder)

    file_path = os.path.join(folder, "financial_report.pdf")
    print(f"Attempting to save PDF to: {file_path}")

    # 2. Setup Document
    doc = SimpleDocTemplate(file_path, pagesize=letter)
    styles = getSampleStyleSheet()
    story = []

    # 3. Add Content
    story.append(Paragraph("FOREIGN EXCHANGE MARKET ANALYSIS REPORT", styles["Title"]))
    story.append(Spacer(1, 12))

    story.append(Paragraph("Executive Summary", styles["Heading2"]))
    
    # Use .get() or direct keys from your .describe() output
    stats_text = (
        f"Average Rate: {summary_stats['closingrate']['mean']:.2f}<br/>"
        f"Max Rate: {summary_stats['closingrate']['max']:.2f}<br/>"
        f"Min Rate: {summary_stats['closingrate']['min']:.2f}<br/>"
        f"Volatility (Std Dev): {summary_stats['closingrate']['std']:.2f}"
    )
    story.append(Paragraph(stats_text, styles["Normal"]))
    story.append(Spacer(1, 12))

    # 4. Add AI Insights
    story.append(Paragraph("Market Analysis & Insights", styles["Heading2"]))
    # Cleaning up text for ReportLab compatibility
    clean_ai_text = ai_text.replace("\n", "<br/>")
    story.append(Paragraph(clean_ai_text, styles["Normal"]))

    # 5. Build PDF
    try:
        doc.build(story)
        print("PDF successfully created in NFEM folder ✅")
    except Exception as e:
        print(f"An error occurred: {e}")

# --- EXECUTION BLOCK ---
# This actually triggers the function using your variables
stats = df[["closingrate", "total_turnover"]].describe().to_dict()
# Extracting the markdown text you wrote in the notebook as a string
insights = """The exchange rate ranges from 1337 to 1660, indicating significant volatility. 
Recent trends show fluctuations around 1380-1390. 
Liquidity observations suggest periods of high market interest followed by zero-trading days."""

create_pdf(stats, insights)
```
